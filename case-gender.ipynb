## Subsets of CASE + GENDER that Cover at least 170 Languages

```python
loc = Locator(170,include=gencase)
loc.main('discard')
```

```python
groups = sorted(loc.flagged,key=lambda g: g.quality_index,reverse=True)
```

### So the 1st PC 

is by and large made up of gender features, while the 2nd PC more from case features.

Taking a different approach, we can check directly which features make for good silhouettes.

```python
stats = list()
ind = list()
for f in gencase:
    part = [g for g in groups if f in g.cols]
    g2 = np.array([g.best_silhouette('genetic',2)[0] for g in part])
    g4 = np.array([g.best_silhouette('genetic',4)[0] for g in part])
    stats.append({
        'gen2 mean': g2.mean(),
        'gen2 std' : g2.std(), 
        'gen4 mean':g4.mean(),
        'gen4 std': g4.std()
    })
    ind.append("{:s} ({:d})".format(f,len(part)))
stats = pd.DataFrame(stats,index=ind)
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.set_title('Average Silhoutte Score (2 Clusters) by Feature')
ax.set_ylabel('Average Silhouette Score')
ax.set_xlabel('Feature (number of sets participated)')
stats['gen2 mean'].plot.bar(yerr=stats['gen2 std'],figsize=(12,7),ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(12,7))
ax.set_title('Average Silhoutte Score (4 Clusters) by Feature')
ax.set_ylabel('Average Silhouette Score')
ax.set_xlabel('Feature (number of sets participated)')
stats['gen4 mean'].plot.bar(yerr=stats['gen4 std'],figsize=(12,7),ax=ax)
```

## So,

It seems safe to say that the features that best separate language families are all the gender-related ones, especially 30A, 31A and 32A, and the case related ones that come from Nominal Categories  -- not Morphology or Simple Clauses -- namely 49A, 50A and 51A.
