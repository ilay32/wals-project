# Groups of Features that Separate Languages by Headness

```python
from locator import *
%matplotlib inline
```

##  Including Word Order

> but keeping the feature sets at most 67%  word order

```python
loc = Locator(900,loi=hlangs,loi_count=50,heterogenous=0.6,cutoff=5)
loc.main()
```

## For Example, The First 2:

```python
for g in loc.flagged[:2]:
    s = g.polinsk_separation(pairwise=True)
    print(g)
    print("separation of headness classes (pairwise average):",s)
    print("\n\n")
```

## Plots for the Best Two:

```python
loc.flagged.sort(key=lambda g: g.polinsk_separation(pairwise=True),reverse=True)
```

```python
loc.flagged[0].plot_polinsk_prop()
```

```python
loc.flagged[1].plot_polinsk_prop()
```
