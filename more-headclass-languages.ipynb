## Maximize Polynski languages (head class)
    
    these are the top quality index sets found with at least 120 Polynski languages, only "A's", with 143A a must

```r
library(FactoMineR)
dat1 <- read.csv('../miscsv/143A-81A-82A-83A.csv')
findall2 <- function(l,df) {
    n <- grep(paste(l,collapse="|"),df$Name)
    g <- which(df$genus %in% l)
    f <- which(df$family %in% l)
    unique(unlist(c(n,g,f)))
}

class1 = findall2(c('Japanese','Korean','Tamil'),dat1)
class2 = findall2(c('German', 'Persian', 'Latin','Tsez', 'Avar','Basque','Germanic'),dat1)
class3 = findall2(c('Tongan','Mayan','Irish'),dat1)
class4 = findall2(c('Indonesian','Yucatec'),dat1)
class5 = findall2(c('English','Russian','Romance','Bantoid'),dat1)

headclass2 <- function(i,df) {

    ans <- 'unknown'
    if(i %in% class1) {
        ans <- 'rigid final'
    }
    if(i %in% class2) {
        ans <- 'non-rigid final'
    }
    if(i %in% class3) {
        ans <- 'clearly initial'
    }
    if(i %in% class4) {
        ans <- 'svo initial'
    }
    if(i %in% class5) {
        ans <- 'svo sundry'
    }
    ans
}
```

```r
y <- sapply(1:1154,headclass2)
dat1 <- cbind(headclass = as.factor(y), dat1)
dat1[0:2,]
```

```r
mca1 <- MCA(dat1,quali.sup=1:12,graph=FALSE)
knownids = unique(unlist(c(class1,class2,class3,class4,class5)))
plot(mca1,choix="ind",habillage="headclass",select=knownids,invisible=c("var","quali.sup"),unselect="1",cex=0.5)
```

## Maximal Number of known verb/noun ratio

```r
dat4 <- read.csv('chunked-feature-sets/143A-143E-143F-143G-144A-81A-82A-83A-88A-89A-799.csv')
dat4$X <- NULL
low <- findall2(c('Halkomelem','Zapotec','Malagasy','Maori','Zinacantec Tzotzil','Irish'),dat4)
medium <- findall2(c('Bahasa','Vietnamese','Hebrew','Swahili','Romanian','Greek','Spanish'),dat4)
high <- findall2(c('Polish','Korean','Japanese','Tamil','Chinese','English',
                 'Czech','German','Basque','Archi','Telugu','Latin','Dutch','Tsez','Hungarian'),dat4)

ratioclass <- function(i) {
    ans = 'unkown'
    if(i %in% low) {
        ans <- 'low'
    }
    if(i %in% medium) {
        ans <- 'medium'
    }
    if(i %in% high) {
        ans <- 'high'
    }
    ans
}

y <- sapply(1:nrow(dat4),ratioclass)
dat4 <- cbind(ratioclass=as.factor(y),dat4)
knownratios <- unique(unlist(c(low,medium,high)))
mca4 <- MCA(dat4,quali.sup=1:11,graph=FALSE)
plot(mca4,choix="ind",habillage="ratioclass",select=knownratios,invisible=c("var","quali.sup"),unselect=1,cex=0.5)
```

```r
dat5 <- read.csv('../wals-project/chunked-feature-sets/143A-143E-143F-143G-144A-81A-82A-83A-89A-886.csv')
dat5$X <- NULL
y <- sapply(1:nrow(dat5),ratioclass)
dat5 <- cbind(ratioclass=as.factor(y),dat5)
knownratios <- unique(unlist(c(low,medium,high)))
mca5 <- MCA(dat5,quali.sup=1:11,graph=FALSE)
plot(mca5,choix="ind",habillage="ratioclass",select=knownratios,invisible=c("var","quali.sup"),unselect=1,cex=0.5)
```
